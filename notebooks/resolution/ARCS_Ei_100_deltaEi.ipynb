## Formula

\begin{align}
\frac{\Delta E_i}{E_i} & = 2 \frac{\Delta v_i}{v_i} \\
\frac{\Delta v_i}{v_i} & = \sqrt{ \big( \frac{\Delta t_i}{t_i} \big)^2 + \big(\frac{\Delta L_i}{L_i} \big)^2}
\end{align}

## Instrument parameters

```python
Ei = 100
Lmod2sample=13.6
sample_size = 10*.01 #
Li = Lmod2fc = 11.61
```

## Calculations

```python
from mcni.utils import conversion as Conv
```

```python
vi = Conv.e2v(Ei)
print vi
```

```python
ti = Li/vi*1e6 # microsecond
print ti
```

```python
dtmod = 10 # microsecond
dtfc = 8 # microsecond
```

```python
dti = np.sqrt(dtmod**2 + dtfc**2)
print dti
```

```python
# this is really small and does not seem to matter much
div = 0.01
dLi_over_Li = div*div
```

```python
dEi_over_Ei = 2*((dti/ti)**2 + dLi_over_Li**2)**.5
```

```python
print dEi_over_Ei
```

## the calculated broadening

```python
dEi = Ei * dEi_over_Ei
print dEi
```
